```python
# Set the path to your video file in Drive
video_path = "/content/drive/MyDrive/player-reidd/15sec_input_720p.mp4"
```

```python
from ultralytics import YOLO
import cv2

# Load the YOLOv8 model (from internet)
model = YOLO('yolov8n.pt')  # Pre-trained lightweight model

# Use the video path from Step 2
cap = cv2.VideoCapture(video_path)

frames = []
for i in range(10):
    ret, frame = cap.read()
    if not ret:
        break
    # Run detection
    results = model(frame)
    # Draw boxes on detected frame
    annotated = results[0].plot()
    # Save the annotated frame as image
    cv2.imwrite(f'frame_{i}.jpg', annotated)
    frames.append(f'frame_{i}.jpg')

cap.release()
print("✅ Detection complete!")
```

```python
from IPython.display import Image, display

for frame in frames:
    display(Image(filename=frame))
```
